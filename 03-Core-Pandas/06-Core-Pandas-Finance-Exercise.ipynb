**Task: Import the necessary libraries**

```python
# CODE HERE
import pandas as pd
import numpy as np
```

**Task: Read the two datasets provided in the current directory (also available in the DATA folder. (The file names are: constituents.csv and constituents-financials.csv)**

```python
# CODE HERE
constituents_df = pd.read_csv('constituents.csv')
constituents_financials_df = pd.read_csv('constituents-financials.csv')
```

**Task: List all columns of the two datasets. Additionally retrieve the datatype for each column**

```python
# CODE HERE
constituents_list = constituents_df.columns.tolist()
constituents_financials_list = constituents_financials_df.columns
print(constituents_financials_list)
```

**Task: Print the first 5 rows of both datasets**

```python
# CODE HERE
display(constituents_df.head(5))
display(constituents_financials_df.head(5))
```

**Task: Drop the SEC Filings column**

```python
# CODE HERE
constituents_financials_df.drop('SEC Filings', axis = 1)
```

**Task: Set the Symbol column to the index in the financials dataset**

```python
# CODE HERE
constituents_financials_df = constituents_financials_df.set_index('Symbol')
constituents_financials_df
```

**Task: What are the 10 largest companies according to market cap?**

```python
# CODE HERE
constituents_financials_df['Market Cap'].nlargest(10)
```

**Task: Return all information about the 10 largest companies according by market cap**

```python
# CODE HERE
constituents_financials_df.nlargest(10,'Market Cap')
```

Note: The difference between GOOG and GOOGL is, that GOOG shares have no voting rights

**Task: Drop GOOG**

```python
# CODE HERE
constituents_financials_df.nlargest(10,'Market Cap')
constituents_financials_df.drop('GOOG', axis = 0)
```

**Task: How many companies have a dividend yield > 4% ?**

```python
# CODE HERE

len(constituents_financials_df[constituents_financials_df['Dividend Yield']>4])
```

**Task: What is the mean Earnings per Share for all companies with a market cap > 1e+11**

```python
# CODE HERE
constituents_financials_df[constituents_financials_df['Market Cap']> 1e+11]['Earnings/Share'].mean()
```

**Task: How many companies have a positive earnings per shares ratio?**

```python
# CODE HERE
(constituents_financials_df['Earnings/Share']>0).sum()
```

**Task: Which company pays the highest dividend yield? What was its 52 week high?**

```python
# CODE HERE
constituents_financials_df['Dividend Yield'].nlargest(1)
constituents_financials_df.sort_values(by="Dividend Yield", ascending=False).iloc[0]
```

**Task: Return the company with the largest spread between 52 weeks high and low**

```python
# CODE HERE
value = abs(constituents_financials_df['52 Week High']-constituents_financials_df['52 Week Low']).idxmax()
constituents_financials_df.loc[value]
```

**Task: Return all companies whith a price between \\$ 50 and \\$ 100**

```python
# CODE HERE
constituents_financials_df[constituents_financials_df['Price'].between(50,100)]
```

**Task: The market cap is really hard to read. Create a new column called "Market Cap in Billion" which shows the market cap in billions**

```python
# CODE HERE
constituents_financials_df['Market Cap in Billion'] = constituents_financials_df['Market Cap']/1000000000000
constituents_financials_df
```

**Task: Is there a correlation between the market cap and the Dividend Yield?**

```python
# CODE HERE
constituents_financials_df[['Market Cap', 'Dividend Yield']].corr()
```

**Task: Merge the financials dataframe with the constituents dataframe**

```python
# CODE HERE
new_df = pd.merge(constituents_df,constituents_financials_df, on = 'Symbol',how = 'inner')
new_df
```

**Task: Print all sectors. How often does each sector occur?**

```python
# CODE HERE
new_df['Sector'].unique()
new_df['Sector'].value_counts()
```

**Task: Replace Information Technology by "IT"**

```python
#CODE HERE
new_df.replace('Information Technology', 'IT')
```

**Task: Add a \$ before the stock price**

```python
# CODE HERE
new_df['Price'] = new_df['Price'].apply(lambda df: '$'+str(df))
```
